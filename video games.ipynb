```python
import pandas as pd
from sqlalchemy import create_engine
```

# Extract CSVs into DataFrames

```python
dataset_file = "resources/Video Games Dataset.csv"
dataset_df = pd.read_csv(dataset_file)
dataset_df.head()
```

```python
sales_file = "resources/vgsales.csv"
sales_df = pd.read_csv(sales_file)
sales_df.head()
```

```python
charts_file = "resources/vgchartz-7_7_2020.csv"
charts_df = pd.read_csv(charts_file)
charts_df.head()
```

```python
developers_file = "resources/video-games-developers.csv"
developers_df = pd.read_csv(developers_file)
developers_df.head()
```

# Transform premise DataFrame

```python
dataset_df.head()
```

```python
cols = ['Name', 'Platform', 'Year', 'Genre', 'Publisher', 'Global_Sales']
dataset_transformed= dataset_df[cols].copy()

dataset_transformed= dataset_transformed.rename(columns={'Name': 'name', 'Platform':'platform', 'Year': 'year',
                                                         'Genre': 'genre', 'Publisher':'publisher', 'Global_Sales':'global_sales'})



dataset_transformed.head()
```

```python
cols2 = ['Name', 'Platform', 'Year', 'Genre', 'Publisher', 'Global_Sales']
sales_transformed= sales_df[cols2].copy()

sales_transformed= sales_transformed.rename(columns={'Name': 'name', 'Platform':'platform', 'Year': 'year',
                                                         'Genre': 'genre', 'Publisher':'publisher', 'Global_Sales':'global_sales'})



sales_transformed.head()
```

```python
cols = ['title', 'console', 'genre', 'publisher', 'developer']
charts_transformed= charts_df[cols].copy()

charts_transformed= charts_transformed.rename(columns={'title': 'name', 'console':'platform',
                                                         'genre': 'genre', 'publisher':'publisher', 'developer':'developer'})



charts_transformed.head()
```
